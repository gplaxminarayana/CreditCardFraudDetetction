# Credit card fraud detection

### Importing the libraries and datasets 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data  =pd.read_csv('creditcard.csv')
```

```python
data
```

```python
data.isna().sum()
```

### Distribution

```python
data.Class.value_counts()
```

#### This dataset is highly unbalanced as the legit transactions are more

```python
legit = data[data['Class']==0]
fraud = data[data['Class']==1]
```

```python
legit.shape ,  fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

### Undersampling the dataset

```python
legit_sample = legit.sample(n=492)
```

### Preparing the new dataset

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset
```

```python
new_dataset.Class.value_counts()
```

```python
#111
import seaborn as sns
```

```python
#sns.pairplot(new_dataset, hue='Class')
```

### Feature engineering

```python
X= new_dataset.drop(columns=['Class'])
y= new_dataset.Class
```

### Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,stratify=y, random_state=2)
```

### Model training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
classifierSVM = SVC()
classifierSVM.fit(X_train ,y_train)
classifierNB=GaussianNB()
classifierNB.fit(X_train,y_train)
classifierRF = RandomForestClassifier(n_estimators=100)
classifierRF.fit(X_train,y_train)
classifierLR = LogisticRegression()
classifierLR.fit(X_train,y_train)
classifierKNN = KNeighborsClassifier(n_neighbors=11)
classifierKNN.fit(X_train,y_train)
```

## Model prediction 


```python
y_predSVM= classifierSVM.predict(X_test)
y_predNB= classifierNB.predict(X_test)
y_predLR= classifierLR.predict(X_test)
y_predRF= classifierRF.predict(X_test)
y_predKNN= classifierKNN.predict(X_test)
```

### Confusion matrix & Accuracy score 

```python
from sklearn.metrics import accuracy_score
```

```python
accSVM= accuracy_score(y_predSVM,y_test)
accNB= accuracy_score(y_predNB,y_test)
accLR= accuracy_score(y_predLR,y_test)
accRF= accuracy_score(y_predRF,y_test)
accKNN= accuracy_score(y_predKNN,y_test)
```

```python
print('Accuracy using Support Vector Machine = ' , accSVM)
print('Accuracy using Naive Bayes = ' , accNB)
print('Accuracy using Logistic Regression = ' , accLR)
print('Accuracy using Random Forest Classifier = ' , accRF)
print('Accuracy using K-Nearest Neighbour = ' , accKNN)
```
